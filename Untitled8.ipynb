#Задача 3
###Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
data = df[['price', 'sqft_living', 'floors', 'condition', 'grade','waterfront', 'view', 'yr_built', 'sqft_lot', 'bedrooms','bathrooms']]
corr_matrix = data.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

Сильное влияние на стоимость оказывают жилая площадь и оценка дома.Чуть в меньшей степень количество ванных комнат и вид из окон. Так же есть небольшая корреляция с этажностью, кол-вом спален и нахождением на набережной. 
Отрицательные кореляции присутствуют у года постройки и состояния. И так же состояние дома не влиет на его этажность.
